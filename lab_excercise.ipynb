```python
import random

print("What is your name?")
user = input()
print("Hello " + user + ", please choose a number between 1 and 20!")


roll = (random.randint(1, 20))


start = 5

while start != 0:
    number = int(input())
    start = start - 1

    if number == roll:
        print("You guessed right!")
        break
    elif number < roll:
        print("The number is higher")
    elif number > roll:
        print("The number is lower")


print("The correct number was " + str(roll))
```
